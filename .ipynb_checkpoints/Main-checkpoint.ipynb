```python
import sys
import nibabel as nib
import matplotlib.pyplot as plt
from scipy.ndimage import zoom
import numpy as np
from glob import glob
from data_loader import test_preprocessing, show_slices
```

```python
paths = glob("/Users/wangqin/Downloads/ADNI-2//**/*.nii", recursive=True)
print(len(paths))
nii_names = list(map(lambda st:st.split('/')[-1], paths))
l = list(map(lambda st:print(st), nii_names))
```

```python
def test_preprocessing(X):
    x,y,z,_ = X.shape
    X = np.reshape(X, (x,y,z))
    X = np.rot90(X, 1, axes=(0,2))
    X = np.rot90(X, 2, axes=(1,2))
    
    return X

epi_imgs = []
for path in paths:
    epi_img = nib.load(path)
    epi_img_data = epi_img.get_fdata()
    epi_imgs.append(epi_img_data)
data1 = epi_imgs[0]
#data1 = test_preprocessing(epi_imgs[0])
x,y,z=data1.shape
print(data1.shape)
show_slices([data1[x//2, :, :],data1[:, y//2, :], data1[:, :, z//2]])
```
